# **_EXTRACCIÓN DE DATOS PARA ML_**

### Se extraen los datos de la pagina web para trabajar el modelo 

```python
# Importamos las librerias a utilizar para el analisis 
import pandas as pd
import requests
import calendar
import io
```

```python
# Se procede a la extraccion de los datos
all_dfs = []

for i in range(1950, 2024):
    year = i
    # Obtener el último día del mes correspondiente
    starttime = f"{year}-01-01"
    endtime = f"{year}-12-31"
    url = f"https://earthquake.usgs.gov/fdsnws/event/1/query?format=csv&starttime={starttime}&endtime={endtime}&minmagnitude=1"
    response = requests.get(url)
    data = response.content
    df = pd.read_csv(io.StringIO(data.decode('utf-8')))
    all_dfs.append(df)    
    print(year)

# Se concatenan los dfs obtenidos en uno solo
result_df = pd.concat(all_dfs, ignore_index=True)
```
